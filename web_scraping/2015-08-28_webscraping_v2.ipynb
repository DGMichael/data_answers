```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://web.archive.org/web/20110514112442/http://unstats.un.org/unsd/demographic/products/socind/education.htm"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content)
```

```python
for row in soup("table"):
    print row
```

```python
#Get the 7th table:
soup("table")[6]
```

```python
A = soup('table')[6].findAll('tr')
```

```python
pop_list = [x for x in A if len(x) == 25] #take only populated elements
```

```python
pop_list[0]
```

```python
len(pop_list)
```

```python
pop_list[0].findAll('td')[4].string
```

```python
import pandas as pd

records = []
for rows in pop_list:
    col=rows.findAll('td')
    #print col
    country = col[0].string
    year = col[1].string
    total = col[4].string
    men = col[7].string
    women = col[10].string
    record = (country,year,total,men, women)
    records.append(record)
    
column_names = ['country','year','total','men','women']

df = pd.DataFrame(records, columns=column_names)
```

```python
df.head()
```

```python
df.set_index('country',inplace = True)
```

```python
#push to a sql database:
import sqlite3 as lite
con = lite.connect('un_data.db')
cur = con.cursor()

with con:
    cur.execute("CREATE TABLE edu_years (country TEXT, year REAL, total REAL, men REAL, women REAL);")
```

```python
#Check indexing
df.ix["Afghanistan"].tolist()
#[0] = year
#[1] = total
#[2] = men
#[3] = women
```

```python
#Check the df indexing:
for country in df.index.tolist()[0:5]:
    print country
```

```python
with con:
    for country in df.index.tolist(): #For each country in the dataframe...
        year,total,men,women = df.ix[country] #Pull the data and unpack into temp variables
        #Setup parameterized sql query:
        query = "INSERT INTO edu_years (country, year, total, men, women) VALUES (?,?,?,?,?)"
        cur.execute(query,(country,year,total,men,women))
```
